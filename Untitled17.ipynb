```python
# fitting training data
def fit(X_train,Y_train):
    """Function to fit he training data into the tree.
                                   
    """

    result = {}
    result["totalData"] = 0
    
    # store names of all classes
    classValues = set(Y_train)
    
    # traversing through each class
    for current in classValues:
        result[current] = {}
        
        # getting the data points which belongs to current class by using a boolean array
        currentClassRow = (Y_train == current)
        
        # storing the features of current data points
        X_train_current = X_train[currentClassRow]
        
        # storing the target class of data points of current
        Y_train_current = Y_train[currentClassRow]
        
        # finding the total number of features
        numFeatures = X_train.shape[1]
        
        # initializing the total count for the current to 0
        result[current]["totalCount"] = 0
        
        # traversing through all features
        for j in range(numFeatures):
            
            # finding the number of time the jth word appeared in current features
            result[current][j] = X_train_current[:,j].sum()
            
            # updating the total_count value of that word
            result[current]["totalCount"] += result[current][j]
        # updating the total number of words in all documents
        result["totalData"] += result[current]["totalCount"]
        
        # check the progress
        print("%",end="")
        

    return result
```

```python
# finding probablity for a specific class
def probability(dictionary, x, current):
    """Function to calculate the probability for the current class

    """
    # avoid the final probablities to approaching zero
    output = np.log(dictionary[current]["totalCount"]) - np.log(dictionary["totalData"])
    
    # finding the total number of features
    numFeatures = len(x)
    
    # traversing through each feature
    for j in range(numFeatures):
        xj = x[j]
        # ignoring the features whose value is zero
        if x[j]==0:
            continue
       
        # Laplace Correction
        current_class_with_xj = dictionary[current][j] + 1
        current_class_total_count = dictionary[current]["totalCount"] + numFeatures
        
        #log probablities
        current_xj_prob = np.log(current_class_with_xj) - np.log(current_class_total_count)
        # adding to the intial output value to get the final value
        output += current_xj_prob
    return output
```

```python

def predictSinglePoint(dictionary,x):
    """ Function to predict values for a single point

    """
    classes = dictionary.keys()
    
    # assuming the initial probablity
    p = -1000
    
    # initialising the variable
    bestClass = -1
    firstRun = True
    
    # traversing through each classes
    for current in classes:
        
        # as one of the keys in dictionary is the total_count to checking if it is then continue
        if (current == "totalData"):
            continue
            
        # get the probality for the current_class
        pCurrent = probability(dictionary, x, current)
        
        # checking if the probablity is best from the previous one
        if (firstRun or pCurrent > p):
            p = pCurrent
            bestClass = current
        firstRun = False

    return bestClass
```

```python
# predicting the classes
def predict(dictionary, X_test):
    """
    Function predict the classes

    """

    Y_pred = []  
    count=1
    
    for x in X_test:
        
        #get the prediction for a single data point
        x_class = predictSinglePoint(dictionary,x)
        Y_pred.append(x_class)
    return Y_pred
```
